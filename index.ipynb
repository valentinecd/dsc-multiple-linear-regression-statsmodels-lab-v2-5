## The Ames Housing Dataset

The [Ames Housing dataset](http://jse.amstat.org/v19n3/decock.pdf) is a newer (2011) replacement for the classic Boston Housing dataset. Each record represents a residential property sale in Ames, Iowa. It contains many different potential predictors and the target variable is `SalePrice`.

````python
import pandas as pd
ames = pd.read_csv("ames.csv", index_col=0)
ames
````

````python
ames.describe()
````

We will focus specifically on a subset of the overall dataset. These features are:

```
LotArea: Lot size in square feet

1stFlrSF: First Floor square feet

GrLivArea: Above grade (ground) living area square feet
```

````python
ames_subset = ames[['LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']].copy()
ames_subset
````

## Step 1: Visualize Relationships Between Features and Target

For each feature in the subset, create a scatter plot that shows the feature on the x-axis and `SalePrice` on the y-axis.

````python
import matplotlib.pyplot as plt
sale_price = ames_subset['SalePrice']
fig, ax = plt.subplots( 1, 3, figsize = (15, 10))

ax[0].scatter(ames_subset['LotArea'], sale_price, alpha = 0.2)
ax[0].set_title("Lot Area")
ax[1].scatter(ames_subset['1stFlrSF'], sale_price, alpha = 0.2 )
ax[1].set_title("1stFlrSF")
ax[2].scatter(ames_subset['GrLivArea'], sale_price, alpha = 0.2)
ax[2].set_title("GrLivArea")

plt.show()
````

````python
# All 3 features seem to have a linear relationship with Sales Price
````

## Step 2: Build a Simple Linear Regression Model

Set the dependent variable (`y`) to be the `SalePrice`, then choose one of the features shown in the subset above to be the baseline independent variable (`X`).

Build a linear regression using StatsModels, describe the overall model performance, and interpret its coefficients.

````python
y = ames_subset['SalePrice']
X_baseline = ames_subset[["GrLivArea"]]
X_baseline.head()
````

````python
# Your code here - import StatsModels, fit baseline model, display results
import statsmodels.api as sm

one_feature_model = sm.OLS(y, sm.add_constant(X_baseline))
one_feature_results = one_feature_model.fit()

print(one_feature_results.summary())
````

````python
"""
R-Squared shows that 50% of the variance in Sales Price is explained the model
pvalue close to 0 indicates a model is statistically significant
constant is 18,600 which indicates that a house with GrLivAre of 0 would sell for $18,600
GrLivArea has a pvalue close to 0 so should be included in the model
     - as the great living area sf increases by 1 sq feet the price increases by $107.13
"""
````

## Step 3: Build a Multiple Linear Regression Model

For this model, use **all of** the features in `ames_subset`.

````python
y = ames_subset['SalePrice']
X = ames_subset.drop('SalePrice', axis=1)
X.head()
````

````python
three_feature_model = sm.OLS(y, sm.add_constant(X))
three_feature_results = three_feature_model.fit()

print(three_feature_results .summary())
````

````python
"""
R-Squared shows that 56.5% of the variance in Sales Price is explained by the model
pvalue close to 0 indicates a good model fit
constant is significant with p_vale < 0.05
Lot area has a pvalue slightly bigger than 0.05 so should not be included in the model
1stFlrSF has a pvalue close to 0 so should be included in the model
    - as the 1st floor sf increases by 1 sq feet the price increase by $60
GrLivArea has a pvalue close to 0 so should be included in the model
     - as the great living area sf increases by 1 feet the price increases by $80.6 sq
"""
````

## Step 4: Create Partial Regression Plots for Features

Using your model from Step 3, visualize each of the features using partial regression plots.

````python
fig = plt.figure(figsize=(15,5))
sm.graphics.plot_partregress_grid(three_feature_results, exog_idx=["LotArea", "1stFlrSF", "GrLivArea"], grid=(1,3), fig=fig)
plt.tight_layout()
plt.show()
````
